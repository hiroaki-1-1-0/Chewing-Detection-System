```python
import matplotlib.pyplot as plt
import numpy as np

error_ratios = [[0.00, 5.00, 5.26, 0.00, 0.00, 0.00, 5.00, 10.53, 5.00, 5.00, 0.00, 0.00],
                [0.00, 0.00, -1, 5.00, 5.00, 5.00, 0.00, 5.00, -1, 5.00, 5.26, -1],
                [5.00, 5.00, -1, 5.00, 5.00, 5.00, 10.00, 20.00, -1, 5.00, 10.00, -1],
                [-1, 0.00, -1, -1, -1, 5.00, 5.26, -1, -1, 5.26, -1, -1],
                [-1, 5.00, 5.00, -1, 5.00, -1, 0.00, 20.00, 5.00, -1, 5.00, 0.00],
                [5.00, 5.00, 0.00, 5.00, 0.00, 5.00, 5.00, 10.53, 0.00, 5.00, 5.00, 5.00],
                [0.00, 5.26, 10.00, 10.00, 5.00, 10.00, 5.00, 5.00, 5.00, 10.00, 10.00, 10.00],
                [5.00, 0.00, 5.00, 5.00, 0.00, 5.00, 0.00, 0.00, 5.00, 0.00, 20.00, 5.26],
                [-1, 5.00, -1, -1, -1, 5.00, 5.00, 0.00, -1, 5.00, -1, -1]]

# 転置
transposed_data = list(zip(*error_ratios))

data_to_plot = []
for row in transposed_data:
    filtered_row = [x for x in row if x != -1]
    data_to_plot.append(filtered_row)

labels = [f"参加者{i+1}" for i in range(len(transposed_data))]

fig, ax = plt.subplots(figsize=(12, 6))
ax.boxplot(data_to_plot, showmeans=True)

ax.set_xticklabels(labels)
plt.xticks(rotation=45, ha='right')

ax.set_xlabel("実験参加者")
ax.set_ylabel("誤差率（％）")

ax.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
