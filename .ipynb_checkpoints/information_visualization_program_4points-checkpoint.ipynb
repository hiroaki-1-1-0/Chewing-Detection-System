```python
# tmp
import pandas as pd
import matplotlib.pyplot as plt

# Define the path where the CSV is stored
file_path = 'data/face_landmarks_20241030_131831.csv'

# Load the CSV file
df = pd.read_csv(file_path)

# Filter the data for the first 200 frames
df_200_frames = df[df['Frame'] < 400]
df_200_frames = df[df['Frame'] >= 200]

# Extract the Frame and Distance columns for plotting
frames_200 = df_200_frames['Frame']
distances_200 = df_200_frames['Distance']

# Create a plot with Frame on the x-axis and Distance on the y-axis for the first 200 frames
plt.figure(figsize=(18, 6))
plt.plot(frames_200, distances_200, marker='o', linestyle='-', color='b')
plt.title('Frame vs Distance (First 200 Frames)')
plt.xlabel('Frame')
plt.ylabel('Distance')
plt.grid(True)

# Display the plot
plt.show()
```
